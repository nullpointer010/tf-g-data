```python
import pandas as pd
import requests
import io

# Umbral mínimo de recall
threshold = 0.55

# Rutas “blob” originales
file_paths = {
    "SPSA":   "https://github.com/nullpointer010/tf-g-data/blob/main/SPSA_evaluation.pkl",
    "L_BFGS_B": "https://github.com/nullpointer010/tf-g-data/blob/main/L_BFGS_B_evaluation2.pkl",
    "COBYLA": "https://github.com/nullpointer010/tf-g-data/blob/main/COBYLA_evaluation3.pkl",
    "ADAM":   "https://github.com/nullpointer010/tf-g-data/blob/main/ADAM_evaluation1.pkl"
}

for name, blob_url in file_paths.items():
    # Convertir a URL “raw”
    raw_url = blob_url.replace(
        "https://github.com/",
        "https://raw.githubusercontent.com/"
    ).replace("/blob/", "/")

    # Leer el pickle desde la URL
    resp = requests.get(raw_url)
    resp.raise_for_status()
    df = pd.read_pickle(io.BytesIO(resp.content))

    # Filtrar y ordenar
    df_filt = df[df["recall"] >= threshold]
    df_ord  = df_filt.sort_values(by="accuracy", ascending=False)

    # Mostrar resultados
    print(f"\n======================================== {name} ========================================")
    print(df_ord.head(10))
    print(f"\n========================================================================================")
```
